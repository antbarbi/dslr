```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv("dataset_train.csv")
houses = data["Hogwarts House"].unique()
courses = pd.DataFrame(data.iloc[:, 6:])

for course in courses:
    plt.figure(figsize=(10, 6))
    course_data = [data[data["Hogwarts House"] == house][course].dropna() for house in houses]
    plt.boxplot(course_data, vert=False, tick_labels=houses)
    plt.title(f'Boxplot of {course} Scores by House')
    plt.xlabel('Score')
    plt.ylabel('House')
    plt.show()
```
