```python
import argparse
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

def parse() -> argparse.Namespace:
    parser = argparse.ArgumentParser()

    parser.add_argument(
        "dataset",
        type=str,
        help="filename to get describe for"
    )
    return parser.parse_args()
```

```python
csv_file = "dataset_train.csv"
# csv_file = "dataset_test.csv"
data = pd.read_csv(csv_file)
```

```python
houses = data["Hogwarts House"].unique()
courses = pd.DataFrame(data.iloc[:, 6:])
for course in courses:
    plt.figure(figsize=(10, 6))
    for house in houses:
        house_data = data[data["Hogwarts House"] == house][course]
        plt.hist(house_data.dropna(), bins=30, alpha=0.5, label=house)  # Drop NaN values if any
    plt.title(f'Histogram of {course} Scores by House')
    plt.xlabel('Score')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()
```

From what we can observe Arithmancy seems to be the most homogeneous score between the four houses
